```python
import random
logo = """
   ____                    _____ _          _   _                 _               
  / ___|_   _  ___  ___ __|_   _| |__   ___| \ | |_   _ _ __ ___ | |__   ___ _ __ 
 | |  _| | | |/ _ \/ __/ __|| | | '_ \ / _ \  \| | | | | '_ ` _ \| '_ \ / _ \ '__|
 | |_| | |_| |  __/\__ \__ \| | | | | |  __/ |\  | |_| | | | | | | |_) |  __/ |   
  \____|\__,_|\___||___/___/|_| |_| |_|\___|_| \_|\__,_|_| |_| |_|_.__/ \___|_|   
                                                                                 
"""

print(logo)

EASY_LEVEL_TURNS = 10
HARD_LEVEL_TURNS = 5

def compare(rnd_number, guess, turns):
    if rnd_number > guess:
        print("Too high")
        print("Guess again")
        return turns-1
    elif rnd_number < guess:
        print("Too low")
        print("Guess again")
        return turns-1
    elif rnd_number == guess:
        print("You win")

def set_difficulty():
    level = input("Choose a difficulty. Type 'easy' or 'hard': ")
    if level == "easy":
        return EASY_LEVEL_TURNS
    else:
        return HARD_LEVEL_TURNS

def game():
    print("Welcome to the number guessing game!")
    print("I'm thinking of a number between 1 and 100.")
    rnd_number = random.randint(1,100)

    # difficulty = input("Choose a difficulty. Type 'easy' or 'hard': ").lower()
    # health_level = 10 if difficulty == "easy" else 5
    did_found = False
    turns = set_difficulty()


    guess = 0
    while guess != rnd_number:
        print(f"You have {turns} attemps reamining to guess the number")
        guess = int(input("Make a guess:"))
        
        turns = compare(rnd_number, guess, turns)
        
        if turns == 0:
            print("You have run out of guesses, you lose.")
            return

game()

```
